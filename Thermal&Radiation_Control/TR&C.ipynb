```python
import math
emmisivity_solar = 0.825
absortivity_solar=  0.805
eta_t = emmisivity_solar
eta_b = emmisivity_solar 
alpha_t = absortivity_solar
alpha_b = absortivity_solar
Volume = 4 * math.pi**2.5 
au = 9.58 
Gs = 1358 
Gs = Gs/ (au**2)
alpha = 0.3 
theta = 0.5*math.pi
Qwmax = 939
Qwmin = 626
sigma = 5.67 * (10**(-8))
eta = 0.8 
Tupper = 45 + 273.15
Tlower = -5 + 273.15
a = 0.73  
q1 = 0.193
H = 100 *(10**3) 
R_enceladus = 252.1 * (10**3)
rho = math.asin(R_enceladus/(H + R_enceladus))
n = 0.1 
```

```python
class thermal_characteristics:
    def __init__(self,Volume,Gs,alpha,theta,Qwmax,Qwmin,sigma,eta,Tupper,Tlower,q1,a,rho):
        self.Volume = Volume
        self.Gs = Gs
        self.alpha = alpha
        self.theta = theta
        self.Qwmax = Qwmax
        self.Qwmin = Qwmin
        self.sigma = sigma
        self.eta = eta
        self.Tupper = Tupper
        self.Tlower = Tlower
        self.q1 = q1
        self.a = a
        self.rho = rho

    def diametercalc(self):
        ribs_length_cube = Volume ** (1/3)
        surface_area_cube = ribs_length_cube * 3
        diameter = math.sqrt(surface_area_cube/math.pi)
        return diameter 
    
    def radiatorarea_hot_cond(self):
        A = Qwmax / ( (Tupper**4)*eta*sigma)
        return A 

    # heat balances 
    def upperTemperature(self):
        Qabsorbed = 0.25* Gs * alpha + 0.5 * q1 * eta * ( 1-math.cos(rho) ) + 0.5 * Gs * a * alpha * ( 1- math.cos(rho) )
        D = self.diametercalc()
        dissipated_energy = Qwmax / (math.pi * (D**2))
        tempmaxlimit = ( (Qabsorbed + dissipated_energy) / (sigma * eta) ) ** 0.25
        return tempmaxlimit
    
    def lowerTemperature(self):
        Qabsorbed = 0.5 * q1 * eta * (1-math.cos(rho)) 
        D = self.diametercalc()
        dissipated_energy = Qwmin/( math.pi * (D**2) )
        tempminlimit = ((Qabsorbed + dissipated_energy)/(sigma * eta)) ** 0.25 
        return tempminlimit
    
    def radiator_low_temp_case(self):
        A = self.radiatorarea_hot_cond()
        Tlower = ( ( (Qwmin/A) ) / ( sigma*eta ) ) ** 0.25
        return Tlower
    
    def temps_solar_array():
        Tmax_A = ((Gs*alpha_t+q1*eta_b*(math.sin(rho)**2)+Gs*a*alpha_b*(math.sin(rho)**2) - n*Gs)/((eta_b+eta_t)*sigma))**0.25
        Tmin_A = ((q1*eta_b*math.sin(rho)**2)/(sigma*(eta_b+eta_t)))**0.25
        return Tmax_A,Tmin_A     
```
